```python
import numpy as np;
import matplotlib.pyplot as plt;
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

# Hill-Clohessy-Wiltshire Equations <br><br>

The equations:<br>
$$x'' = 3n^2x + 2ny'$$
$$y'' = -2nx'$$ 
$$z'' = -n^2z$$

<br><br>

Closed form equations:
<br>
$$x(t) = (4-3cos(nt))x_0 + \frac{sin(nt)}{n}+\frac{2}{n}(1-cos(nt))y'_0$$
$$y(t) = 6(sin(nt)-nt)x_0 + y_0 - \frac{2}{n}(1-cos(nt))x'_0+\frac{4sin(nt)-3nt}{n}y'_0$$
    $$z(t) = z'_0cos(nt) + \frac{z'_0}{n}sin(nt)$$

```python
def hcw_pos(t, init_pos, init_vel, orbital_rate):
    x0,y0,z0 = init_pos
    vx0, vy0, vz0 = init_vel
    
    orbital_t = orbital_rate * t #nt for calculations
    
    xt = (4-3*np.cos(orbital_t))*x0 + np.sin(orbital_t)/orbital_rate*vx0 
    xt += 2*(1-np.cos(orbital_t))*vy0/orbital_rate
    
    yt = 6*(np.sin(orbital_t) - orbital_t)*x0 + y0 - 2*(1-np.cos(orbital_t))*vx0/orbital_rate 
    yt += (4*np.sin(orbital_t) - 3*orbital_t)/orbital_rate*vy0
    
    zt = z0*np.cos(orbital_t) + vz0/orbital_rate * np.sin(orbital_t)
    
    pos_vector = np.vstack((xt,yt,zt))
    return pos_vector
```

```python
x0 = -1
y0 = 0
z0 = 0

vx0 = 0
vy0 = 0
vz0 = 0

orbital_rate = 1

t = np.linspace(0,1,20)

pos_vector = hcw_pos(t,(x0,y0,z0), (vx0,vy0,vz0), orbital_rate)

fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(pos_vector[0,:], pos_vector[1,:], pos_vector[2,:],s=10, color='red') 
ax.plot(pos_vector[0,:], pos_vector[1,:], pos_vector[2,:]) 
ax.scatter(0,0,0,s=200,color='green')
plt.show()
```
